```python
#Accept a list of students and their marks as tuples. 
#Display the name of the student with the highest marks.


#Take input from user
students = eval(input("Enter list of students and marks: "))
print("Enter list of students and marks: ",students)
```

```python
#Find topper using max() and lambda
topper = max(students, key=lambda x: x[1])
print("Topper:", topper[0], f"({topper[1]})")
```
